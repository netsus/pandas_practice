```python
import pandas as pd
from pydataset import data
```

```python
data('grouseticks',show_doc=True)
```

```python
df = data('grouseticks')
```

```python
df
```

```python
print(sorted(df.HEIGHT.unique()))
```

```python
interest_li = [405,450,490,525] # 관심 리스트
```

```python
df_sort = df[df['관심 컬럼'].isin(interest_li)].copy() # 먼저, 데이터 프레임에서 관심값 추출
sorterIndex = dict(zip(interest_li, range(len(interest_li)))) # 정렬을 위한 dictionary 생성
df_sort['sorter'] = df_sort['관심 컬럼'].map(sorterIndex) #해당 dictionary를 이용 정렬을 위해 'sorter'열 생성
df_sort.sort_values('sorter',inplace=True) # sorter를 기준으로 정렬
df_sort.drop('sorter', 1, inplace = True) # sorter 열 삭제
```

```python
height_li = [405,450,490,525] # 관심있는 리스트

df_sort = df[df['HEIGHT'].isin(height_li)].copy()
sorterIndex = dict(zip(height_li, range(len(height_li))))
df_sort['sorter'] = df_sort['HEIGHT'].map(sorterIndex)
df_sort.sort_values('sorter',inplace=True)
df_sort.drop('sorter', 1, inplace = True)
```
